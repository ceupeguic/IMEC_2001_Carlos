## 3. Matemática Básica en Python

## Ejercicio 5: Iteradores y Condicionales

Se tienen el conjunto de información de los meses y la respectiva cantidad de días en una estructura `list`. 

<div class="alert alert-block alert-warning">
    
 <ol>
  <li>Escriba un algoritmo que imprima un diccionario que contiene los nombres de los meses con menos de 31 días y el respectivo número de días.</li>
  <li>Escriba un algoritmo que imprima una lista que contiene los nombres de los meses con menos de 31 días y el respectivo número de días.</li>
</ol>
</div>

___
**Ejemplo** <br>
<font color="#FF5733">Output</font>
```python
{'Febrero': 28,
 'Abril': 30,
 'Junio': 30,
 'Septiembre': 30,
 'Noviembre': 30}

["Febrero - 28", "Abril - 30", "Junio - 30", "Septiembre - 30", "Noviembre - 30"]
```
___

````python
#Lo primero que hago es crear un diccionario con los meses y la cantidad de dias
meses_y_dias = {
    'Enero': 31,
    'Febrero': 28,
    'Marzo': 31,
    'Abril': 30,
    'Mayo': 31,
    'Junio': 30,
    'Julio': 31,
    'Agosto': 31,
    'Septiembre': 30,
    'Octubre': 31,
    'Noviembre': 30,
    'Diciembre': 31}
#Ahora creo el diccionario con los meses que tengan menos de 31 dias y cuantos dias contiene el mes
meses_dic = {mes: dias for mes, dias in meses_y_dias.items() if dias < 31}
print("Diccionario:")
print(meses_dic)
#Ahora creo la lista con los meses que tengan menos de 31 dias y cuantos dias contiene el mes
meses_lista = [(mes, dias) for mes, dias in meses_y_dias.items() if dias < 31]
print("Lista:")
print(meses_lista)
````

### Punto 1

Se busca preparar la nómina de una pequeña empresa. Los datos suministrados son el número de horas trabajadas (`num_horas`) y la tasa de pago (`tasa_pago`). Para esto, se tienen las siguientes condiciones:

- Si `num_horas` es menor o igual a 40, calculamos el pago regular (`pago_regular`) multiplicando `num_horas` por `tasa_pago`, y el pago de horas extras (`pago_extras`) es igual a 0.


- Si `num_horas` es mayor a 40, calculamos el pago regular multiplicando 40 por `tasa_pago` y el pago de horas extras así: $(num\_horas - 40) · tasa\_pago · 1.5$. 

El salario neto (`neto`) se calcula sumando el pago regular (`pago_regular`) y el pago de horas extras (`pago_extras`).

<div class="alert alert-block alert-warning">
    
Escriba un algoritmo que permita calcular el pago de nómina de la pequeña empresa.

</div>

___
**Ejemplo** <br>
<font color="#FF5733">Input</font>
```python
horas_trabajadas = 30
pago_por_hora = 25
```
<font color="#FF5733">Output</font>
```python
neto = 750
```
___

````python
# Primero le pedimos el input al usuario del numero de horas y la tasa de pago por hora
num_horas = int(input("Ingrese el número de horas trabajadas: "))
tasa_pago = float(input("Ingrese la tasa de pago por hora: "))

# Ahora hago el cálculo del pago regular y de horas extras
if num_horas <= 40:
    pago_regular = num_horas * tasa_pago
    pago_extras = 0
else:
    pago_regular = 40 * tasa_pago
    pago_extras = (num_horas - 40) * tasa_pago * 1.5

# Despues calculo el salario neto con las variables anteriormente definidas
ans = pago_regular + pago_extras

# Por ultimo imprimimos los resultados el enunciado pide el salario neto pero se complementa con mas datos
print("El pago regular es:", pago_regular)
print("El pago de horas extras es:", pago_extras)
print("El salario neto es:", ans)
````

<div class="alert alert-block alert-warning">

Cree una función `func` que reciba de entrada la función `f` (definida a continuación) y cuatro números como parámetros. Esta función `func` debe ejecutar la función `f` dos veces (la primera con los dos primeros números, y la última con los dos números finales). Con los dos resultados (i.e., al ejecutar dos veces y por separado la función `f`), calcular la **raíz de la suma de los cuadrados**. Este valor se debe imprimir en pantalla (i.e., `print()`) y retornar (i.e., `return`).

</div>

```python
def f(x, y):
    return x + y    
```

___
**Ejemplo** <br>
<font color="#FF5733">Input</font>
```python
num_1 = 1
num_2 = 2
num_3 = 3
num_4 = 4
```
<font color="#FF5733">Output</font>
```python
rta = 7.615773105863909
```
___


````python
import math

def f(x, y):
    return x + y
# Primero definimos la funcion
def func(f, num1, num2, num3, num4):
    resultao1 = f(num1, num2)
    resultado2 = f(num3, num4)
    
    suma = resultao1**2 + resultado2**2
    raiz = math.sqrt(suma)
    
    return raiz

# Luego le pedimos ingresar los valores a el usuario
num1 = float(input("Ingrese el primer número: "))
num2 = float(input("Ingrese el segundo número: "))
num3 = float(input("Ingrese el tercer número: "))
num4 = float(input("Ingrese el cuarto número: "))

# Por lutimo imprimimos la respuesta
ans = func(f, num1, num2, num3, num4)
print("La respuesta es ", ans)
````
